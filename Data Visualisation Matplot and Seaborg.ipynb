```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Plotting simple line graph
squares = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(squares)
plt.show()
```

```python
#changig the label type and line thickness
squares = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(squares)
ax.plot(squares,linewidth=3)
#set chart title and label axes
ax.set_title("Square numbers",fontsize=24)
ax.set_xlabel("value", fontsize =14)
ax.set_ylabel("square of value",fontsize=14)
#set size of tick label
ax.tick_params(axis='both',labelsize=14)
plt.show()
```

```python
#line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.show()
```

```python
#Bar plot
cate = ['a','d','r','t','y']
value = [13,12,10,23,8]

plt.bar(cate,value)
plt.xlabel('alphabets')
plt.ylabel('num value')
plt.title('barplot')
plt.show()
```

```python
#scatter plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scater plot')
plt.show()
```

```python
#Histogarm
data = sns.load_dataset("tips")
sns.histplot(data['total_bill'],kde = True)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title("histogarm with kde")
plt.show()
```

```python
#pair plot
data = sns.load_dataset("iris")
sns.pairplot(data,hue= 'species')

plt.title("pair plot of iris dataset")
plt.show()
```

```python
#Heatmap
# Load a sample dataset
data = sns.load_dataset("flights")

# Create a pivot table
pivot_data = data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(pivot_data, cmap="YlGnBu", annot=True, fmt="d")
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count Heatmap')
plt.show()
```

```python
#Creat world map
import cartopy.crs as ccrs

# Create a figure and axis with a PlateCarree projection
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})

# Draw coastlines
ax.coastlines()

# Add title
ax.set_title('World Map')

# Show the map
plt.show()
```

```python
# color charts
# Choose a color palette
color_palette = sns.color_palette("Set2", 8)

# Create a bar chart with different colors
x = ["Category 1", "Category 2", "Category 3", "Category 4"]
y = [5, 10, 15, 20]

plt.bar(x, y, color=color_palette)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Colorful Bar Chart')

# Show the chart
plt.show()
```

```python
#Pie chart
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Box plot
import numpy as np

data = np.random.randn(100)
plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

```python
#Histogram with multiple bins
data = np.random.randn(1000)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with 20 Bins')
plt.show()
```

```python
#violin plot
data = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=data)
plt.title('Violin Plot')
plt.show()
```

```python
#pair plot with regression line
data = sns.load_dataset("iris")
sns.pairplot(data, hue='species', kind='reg')
plt.title('Pair Plot with Regression Lines')
plt.show()
```
