```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Plotting simple line graph
squares = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(squares)
plt.show()
```

```python
#changig the label type and line thickness
squares = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(squares)
ax.plot(squares,linewidth=3)
#set chart title and label axes
ax.set_title("Square numbers",fontsize=24)
ax.set_xlabel("value", fontsize =14)
ax.set_ylabel("square of value",fontsize=14)
#set size of tick label
ax.tick_params(axis='both',labelsize=14)
plt.show()
```

```python
#line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.show()
```

```python
#Bar plot
cate = ['a','d','r','t','y']
value = [13,12,10,23,8]

plt.bar(cate,value)
plt.xlabel('alphabets')
plt.ylabel('num value')
plt.title('barplot')
plt.show()
```

```python
#scatter plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scater plot')
plt.show()
```

```python
#Histogarm
data = sns.load_dataset("tips")
sns.histplot(data['total_bill'],kde = True)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title("histogarm with kde")
plt.show()
```

```python
#pair plot
data = sns.load_dataset("iris")
sns.pairplot(data,hue= 'species')

plt.title("pair plot of iris dataset")
plt.show()
```

```python
#Heatmap
# Load a sample dataset
data = sns.load_dataset("flights")

# Create a pivot table
pivot_data = data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(pivot_data, cmap="YlGnBu", annot=True, fmt="d")
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count Heatmap')
plt.show()
```

```python
#Creat world map
import cartopy.crs as ccrs

# Create a figure and axis with a PlateCarree projection
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})

# Draw coastlines
ax.coastlines()

# Add title
ax.set_title('World Map')

# Show the map
plt.show()
```

```python
# color charts
# Choose a color palette
color_palette = sns.color_palette("Set2", 8)

# Create a bar chart with different colors
x = ["Category 1", "Category 2", "Category 3", "Category 4"]
y = [5, 10, 15, 20]

plt.bar(x, y, color=color_palette)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Colorful Bar Chart')

# Show the chart
plt.show()
```

```python
#Pie chart
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Box plot
import numpy as np

data = np.random.randn(100)
plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

```python
#Histogram with multiple bins
data = np.random.randn(1000)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with 20 Bins')
plt.show()
```

```python
#violin plot
data = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=data)
plt.title('Violin Plot')
plt.show()
```

```python
#pair plot with regression line
data = sns.load_dataset("iris")
sns.pairplot(data, hue='species', kind='reg')
plt.title('Pair Plot with Regression Lines')
plt.show()
```

```python
#strip plot
data = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=data, jitter=True)
plt.title('Strip Plot')
plt.show()
```

```python
# Area Chart
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 18, 20]
y2 = [5, 10, 8, 12, 14]

plt.fill_between(x,y1,y2,alpha =0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Chart Example')
plt.show()
```

```python
#Radar chart(Spider chart)
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [4, 5, 2, 7]

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
values += values[:1]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, alpha=0.25)
plt.xticks(angles[:-1], categories)
plt.title('Radar Chart Example')
plt.show()
```

```python
#stacked bar chart
categories = ['Category A', 'Category B', 'Category C']
values1 = [10, 15, 20]
values2 = [5, 10, 15]

plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, label='Group 2', bottom=values1)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')
plt.legend()
plt.show()
```

```python
#Water fall chart
categories = ['Start', 'Step 1', 'Step 2', 'Step 3', 'End']
values = [100, -20, 30, -10, 90]

plt.bar(categories, values, color=['blue', 'red', 'green', 'red', 'blue'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Waterfall Chart Example')
plt.show()
```

```python
#Bubble chart
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Chart Example')
plt.show()
```
